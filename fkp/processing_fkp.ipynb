```python
import pandas as pd
import os
import json
import xlrd
import datetime 
import locale
from openpyxl import load_workbook
```

```python
# ganti ini ya, sesuai dengan kebutuhan kalian
GLOBAL_VAR = {
    "kode_satker":"1507",
    "jadwal_fkp":"Data Jumlah FKP per desa.xlsx",
    "sheet_name":"Sheet5",
    "output_filename": 'Jadwal FKP.xlsx'
}
```

```python
folder = "data"
fkpData = pd.read_excel(folder+"/"+GLOBAL_VAR["jadwal_fkp"], sheet_name=GLOBAL_VAR["sheet_name"])
```

```python
fkpData
```

```python
# data yang dihasilkan di output file nanti, silahkan tambahkan atau kurangin
kolomOutput = [
    "Kecamatan",
    "Desa/Kelurahan",
    "FKP",
    "Tanggal Pelaksanaan",
    "Jumlah Perwakilan dari SLS"
    ]

fkpFinal = []

for index, row in fkpData.iterrows():
    for i in range(1,row['Jumlah FKP']+1):
        result = {}
        result["Kecamatan"] = row["Kecamatan"]
        result["Desa/Kelurahan"] = row["Desa/Kelurahan"]
        result["FKP"]="FKP"+" "+str(i)
        result["Tanggal Pelaksanaan"] = ""
        result["Jumlah Perwakilan dari SLS"]=""
        fkpFinal.append(result)
```

```python
output_folder = 'output'
jsonObj = json.dumps(fkpFinal)
fkpFinalPd = pd.read_json(jsonObj)

path = folder+"/"+GLOBAL_VAR["jadwal_fkp"]
book = load_workbook(path)
writer = pd.ExcelWriter(path, engine = 'openpyxl')
writer.book = book

fkpFinalPd.to_excel(writer, sheet_name = 'Jadwal FKP', index=False)
fkpData.to_excel(writer, sheet_name = 'Data Jumlah FKP per Desa', index=False)
writer.close()
```
